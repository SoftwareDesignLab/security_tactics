```
import TacticClassifier as tc
```

```
tactics = ["authenticate"]
tactics = ["authorize", "data integrity"]
tactics = ["encrypt decrypt", "secure communication"]

experiment = "2"

tactics = ["authenticate", "authorize","data integrity"]
tactics = [ "encrypt decrypt","secure communication"]
experiment = "6"

for tactic in tactics:
  classifier = tc.TacticClassifier(tactic, experiment)
  hyper_parameters = dict()
  hyper_parameters['epochs'] = 3
  hyper_parameters['batch_size'] = 8
  hyper_parameters['learning_rate'] = 2e-5

  hidden_layer_parameters = dict()
  hidden_layer_parameters['use_hidden_layer'] = False
  hidden_layer_parameters['number_of_hidden_units'] = 0

  mean_pooling_parameters = dict()
  #this should be decided based on the length of the samples. 
  #each sequence will be of 510 token(2 tokens reserved for special tokens)
  #Select the number of sequences such that maximum portion of the input is 
  #considered for majority of the samples
  mean_pooling_parameters['use_mean_pooling'] = True
  mean_pooling_parameters['max_sequences'] = 2
  classifier.run_experiment(hyper_parameters, 
                          hidden_layer_parameters, 
                          mean_pooling_parameters)
```
